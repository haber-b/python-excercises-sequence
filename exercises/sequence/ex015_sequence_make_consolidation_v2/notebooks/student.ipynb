## Exercise 1 — Party Supply Calculator

You are planning a party and need to work out how many goodie bags you can make.

### What to do
Write a program that calculates how many full goodie bags you can make, and how many sweets will be left over.

### Variables you will need
Create these variables **in this exact order**:
1. `total_sweets` — stores the number 50 (an `int`)
2. `sweets_per_bag` — stores the number 7 (an `int`)

### Algorithm (step by step)
Follow these steps in your code:
1. Store the number 50 in a variable called `total_sweets`.
2. Store the number 7 in a variable called `sweets_per_bag`.
3. Work out how many full bags you can make using floor division (`//`). Store the answer in a variable called `full_bags`.
4. Work out how many sweets are left over using the modulus operator (`%`). Store the answer in a variable called `sweets_left`.
5. Print exactly this message using an f-string:

`You can make 7 full bags with 1 sweets left over.`

*(Hint: Your f-string should use the variables `full_bags` and `sweets_left`, not the numbers 7 and 1 directly.)*

```python
# Exercise 1 — Party Supply Calculator
# Follow the instructions above step by step.

# 1. Store total_sweets
total_sweets = 50
total_sweets = int(total_sweets)
# 2. Store sweets_per_bag
sweets_per_bag = 7
sweets_per_bag = int(sweets_per_bag)
# 3. Calculate full_bags using //
full_bags = total_sweets // sweets_per_bag
# 4. Calculate sweets_left using %
sweets_left = total_sweets % sweets_per_bag
# 5. Print the result using an f-string
print(f"You can make {full_bags} full bags with {sweets_left} sweets left over.")

# Write your code below:
```

## Exercise 2 — Paint Coverage Calculator

A single can of paint covers 12 square metres. You want to paint a wall that is 5 metres wide and 3 metres tall.

### What to do
Write a program that calculates:
- the area of the wall
- how many cans of paint you need (you can only buy whole cans)
- how much paint will be left unused from the last can

### Variables you will need
Create these variables **in this order**:
1. `width` = 5
2. `height` = 3
3. `coverage_per_can` = 12
4. `area` — width multiplied by height
5. `cans_needed` — use ceiling division logic (hint: if `area` does not divide evenly by `coverage_per_can`, you need one more can). Use integer arithmetic: `(area + coverage_per_can - 1) // coverage_per_can`
6. `paint_left` — calculate the unused coverage: `(cans_needed * coverage_per_can) - area`

Print exactly these two lines:
`Wall area: 15 square metres`
`You need 2 cans; 9 square metres of paint will be unused.`

```python
# Exercise 2 — Paint Coverage Calculator
# Use the formulas from the instructions to calculate and print.
# Write your code below:
width = 5
height = 3
coverage_per_can = 12

width = int(width)
height = int(height)
coverage_per_can = int(coverage_per_can)

area = width * height
cans_needed = (area + coverage_per_can - 1 )// coverage_per_can
paint_left = (cans_needed * coverage_per_can) - area

print(f"Wall area: {area} square metres")
print(f"You need {cans_needed} cans; {paint_left} square metres of paint will be unused.")
```

## Exercise 4 — Pizza Sharing

Write a program that asks for the number of pizzas ordered and the number of people sharing.
Each pizza has 8 slices.

### What to do
1. Ask for the number of pizzas and convert to an integer.
2. Ask for the number of people and convert to an integer.
3. Work out the total slices (`pizzas * 8`).
4. Work out the slices per person using floor division, and the leftover slices using modulus.
5. Print exactly:

`If you ordered 3 pizzas for 5 people:`
`Each person gets 4 slices, with 4 slices left over.`

*(Use an f-string with the variables, not the numbers 3, 5, 4, 4 directly.)*

```python
# Exercise 4 — Pizza Sharing
# Ask for two inputs, calculate, and print the result.
# Write your code below:
number_pizzas = input("Enter number of pizzas: ")
number_pizzas = int(number_pizzas)

number_people = input("Enter number of people: ")
number_people = int(number_people)

total_slices = number_pizzas * 8

slices_each = total_slices // number_people
leftover_slices = total_slices % number_people
print(f"If you ordered {number_pizzas} pizzas for {number_people} people: Each person gets {slices_each} slices, with {leftover_slices} left over.")
```

## Exercise 7 — Savings Goal

Write a program that asks for:
- a savings goal in pounds (as a whole number)
- how much you can save each week in pounds (as a whole number)

Work out:
- how many full weeks it will take to reach the goal (floor division)
- how much money will be left short (modulus)

Print exactly:

`Saving £50 per week towards a £350 goal:`
`It will take 7 full weeks, with £0 left to save.`

*(Remember: a full week means you save the full weekly amount. The "left to save" is the remainder.)*

```python
# Exercise 7 — Savings Goal
# Ask for two inputs, calculate weeks and remainder.
# Write your code below:

savings_goal = input("What is your savings goal?")
savings_goal = int(savings_goal)

goal_per_week = input("How much saved per week?")
goal_per_week = int(goal_per_week)

weeks_taken = savings_goal // goal_per_week
weeks_taken = int(weeks_taken)

amount_left = savings_goal % goal_per_week
amount_left = int(amount_left)

print(f"Saving £{goal_per_week} per week towards a £{savings_goal} goal : It wil take {weeks_taken} full weeks, with £{amount_left} left to save.")


```

## Exercise 8 — Garden Area

Write a program that asks for:
- the side length of a square garden in metres (you can assume whole metres)

Work out:
- the area of the garden (side length squared)
- the side length of a new garden that has twice the area (square root of `area * 2`, rounded to 2 decimal places)

Print exactly:

`A square garden with side 5m has area 25 square metres.`
`A garden with double the area would have side 7.07m.`

*(Hint: Use `** 2` for squaring and `** 0.5` for square root.)*

```python
# Exercise 8 — Garden Area
# Use ** for squaring and square roots.
# Write your code below:

side_length = input("What is the side length of the garden?")
side_length = int(side_length)

area_of_garden = side_length ** 2

double_garden = area_of_garden * 2
double_garden = float(double_garden)

double_side = round(double_garden ** 0.5, 2)
double_side = float(double_side)

print(f"A square garden with a side of {side_length}m has an area of {area_of_garden} square metres. A garden with double the area would have a side length of {double_side}m.")
```

## Exercise 9 — Bill Splitter with Tip

Write a program that asks for:
- the total bill amount in pounds (as a decimal)
- the tip percentage (as a whole number, e.g. 10 means 10%)
- the number of people sharing the bill

Work out:
- the tip amount (bill * tip_percent / 100)
- the total including tip
- the amount each person pays (total including tip divided by number of people)

Round the final per-person amount to 2 decimal places.

Print exactly:

`Total bill: £50.00`
`Tip (10%): £5.00`
`Total with tip: £55.00`
`Each of 4 people pays: £13.75`

*(Use f-strings with `:.2f` to show 2 decimal places.)*

```python
# Exercise 9 — Bill Splitter with Tip
# Three inputs, multiple calculations, formatted output.
# Write your code below:
bill = input("Input the total bill: ")
bill = float(bill)

tip_percentage = input("Input the tip percentage: ")
tip_percentage = int(tip_percentage)

number_splitting_bill = input("Input the amount of people splitting the bill: ")
number_splitting_bill = int(number_splitting_bill)

tip_amount = bill * tip_percentage / 100

total_bill = bill + tip_amount
total_bill = float(total_bill)
total_bill = round(total_bill, 2)

amount_per_person = total_bill / number_splitting_bill
amount_per_person = float(amount_per_person)
amount_per_person = round(amount_per_person, 2)

print(f"Total bill: £{bill} Tip ({tip_percentage}%): £{tip_amount} Total with tip: £{total_bill} Each of {number_splitting_bill} pays: £{amount_per_person}")



```

## Exercise 10 — Classroom Supplies Order

A teacher needs to order supplies for a classroom. Write a program that asks for:
- the number of students
- the number of pencils each student needs
- the number of erasers each student needs
- pencils come in packs of 24
- erasers come in packs of 15

Calculate:
- total pencils needed (students * pencils per student)
- total erasers needed (students * erasers per student)
- packs of pencils to order (remember you can only order whole packs — if you don't have enough, order one more)
- packs of erasers to order (same logic)
- individual pencils left over after opening the packs (if any)
- individual erasers left over after opening the packs (if any)

Print exactly:

`For 30 students:`
`Each student needs 3 pencils and 2 erasers.`
`Order 4 packs of pencils (24 per pack) — 6 pencils will be left over.`
`Order 4 packs of erasers (15 per pack) — 0 erasers will be left over.`

*(Remember: use integer arithmetic. If total_pencils / 24 is not exact, you need `(total_pencils + 24 - 1) // 24` packs.)*

```python
# Exercise 10 — Classroom Supplies Order
# Plan your variables and calculations carefully.
# Write your code below:
students = input("How many students? ")
students_int = int(students)
num_pencils = input("How many pencils per student? ")
num_pencils_int = int(num_pencils)
num_erasers = input("How many erasers per student? ")
num_erasers_int = int(num_erasers)

total_pencils = students_int * num_pencils_int
total_pencils_int = int(total_pencils)
total_erasers = students_int * num_erasers_int
total_erasers_int = int(total_erasers)

pencil_order = (total_pencils_int + 24 - 1) // 24
eraser_order = total_erasers_int / 15

pencil_leftover = (total_pencils_int + 24 - 1) % 24
eraser_leftover = total_erasers_int % 15

print(f"Pencil Pack : {pencil_order}, {pencil_leftover} Eraser Pack : {eraser_order}, {eraser_leftover}")
print(f"For {students} students: Each student needs {num_pencils} pencils and {num_erasers} erasers. Order {pencil_order} packs of pencils (24 per pack) - {pencil_leftover} pencils will be leftover.")
```
