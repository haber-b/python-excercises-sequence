# Sequence Modify F-Strings

In this notebook you practise changing working sequence programs so they use f-strings to build the final output.

## What you'll practise
- Converting `print()` statements with `+` concatenation into f-strings.
- Editing the text and variables inside an f-string.
- Using `input()`, arithmetic, and casting inside short sequence programs.

## Worked examples
```python
name = "Ava"
print(f"Hello, {name}!")
```

```python
score = 7
print(f"Score: {score}")
```

In an f-string, add `f` before the opening quote and place variable names inside `{}`.
You do not need `str()` when a number is already inside an f-string.

## How to work
1. Read the task and expected output.
2. Change the working code so the output matches exactly.
3. Keep the code in sequence only: no `if`, loops, or functions.
4. Use f-strings whenever the task asks for them.
5. From the repository root, run `uv run pytest -q exercises/sequence/ex009_sequence_modify_fstrings/tests/test_ex009_sequence_modify_fstrings.py`.
